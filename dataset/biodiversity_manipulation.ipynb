```python
import pandas as pd
```

```python
df = pd.read_csv("biodiversity_dataset.csv")
df.head()
```

```python
df = df.drop(['datasetKey','infraspecificEpithet'],axis=1)
print(df.head())
```

```python
df = df.drop([' gbifID','occurrenceID'],axis=1)
print(df.head())
```

```python
df = df.drop(['identifiedBy','dateIdentified'],axis=1)
print(df.head())
```

```python
df = df.drop(['catalogNumber','recordNumber','license','rightsHolder'],axis=1)
print(df.head())
```

```python
df = df.drop(['speciesKey','basisOfRecord','institutionCode','collectionCode','recordedBy','typeStatus'],axis=1)
print(df.head())
```

```python
df = df.drop(['depthAccuracy','eventDate','day','month','year','taxonKey','establishmentMeans','lastInterpreted'],axis=1)
print(df.head())
```

```python
df = df.drop(['publishingOrgKey','decimalLatitude','decimalLongitude'],axis=1)
print(df.head())
```

```python
print(df.columns.values)
```

```python
df = df.drop(['coordinateUncertaintyInMeters','coordinatePrecision','elevation','elevationAccuracy','depth','mediaType','issue'],axis=1)
print(df.head())
```

```python
df = df.drop(['countryCode','locality','stateProvince','occurrenceStatus','individualCount'],axis=1)
print(df.head())
```

```python
df = df.drop(['verbatimScientificNameAuthorship'],axis=1)
print(df.head())
```

```python
df.to_csv("modified_biodiversity_dataset.csv")
```
